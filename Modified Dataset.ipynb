```python
import pandas as pd
```

```python
# Read the original dataset
data=pd.read_excel('chatbot_data.xlsx')
data
```

```python
# splitting column "Calendar Quarter"
new = data["Calendar Quarter"].str.split(" ", n = 1, expand = True) 
data["Quarter"]= new[0] 
data["Year"]= new[1] 
```

```python
data
```

```python
# drop some columns
data.drop(columns =["Calendar Quarter","Modified Molecule","Standard Units","DoT Factor"], inplace = True) 
data 
```

```python
data=data[['Country','Quarter','Year','Local Product','US$ MNF']]
```

```python
# lowercasing everything
for i in data.columns:
    if(data[i].dtype!='int64'):
        data[i]=data[i].apply(lambda x: x.lower())
data
```

```python
# save the modified csv file
data.to_csv('dataset.csv',index=False)
```
